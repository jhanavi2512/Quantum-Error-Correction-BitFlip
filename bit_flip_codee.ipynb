```python
#importing required modules from qiskit#
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
#creating quantum circuit#
qc=QuantumCircuit(3,2)
```

```python
#encoding one qubit into three qubit#
qc.h(0)
qc.cx(0,1)
qc.cx(0,2)
```

```python
# introducing bit flip error#
qc.x(1)
```

```python
#error detection#
qc.cx(0,1)
qc.cx(0,2)
```

```python
#measurmrnt#
qc.measure([1,2],[0,1])
```

```python
#drawing circuit#
qc.draw()
```

```python
#introducing simulation#
simulator = AerSimulator()
result = simulator.run(qc, shots=1000).result()
counts = result.get_counts()
```

```python
#shows result#
print(counts)
```

```python
#ploting histogram#
plot_histogram(counts)
```

```python
Result Interpretation

The simulator produced {'01': 1000}.  
This indicates that the bit-flip error occurred on qubit 1.

The error detection circuit successfully identified the flipped qubit.
```
